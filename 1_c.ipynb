#Métodos numéricos de aproximación de la integral



##Método punto medio

```python
import numpy as np

def midpoint(f, a, b, n):
    h = float(b-a)/n
    result = 0
    for i in range(n):
        result += f(a + (i - 0.5)*h)
    result *= h
    return result
```

##Método trapecio

```python
import numpy as np


def trapz(f,a,b,N):

    x = np.linspace(a,b,N+1) # N+1 points make N subintervals
    y = f(x)
    y_right = y[1:] # right endpoints
    y_left = y[:-1] # left endpoints
    dx = (b - a)/N
    T = (dx/2) * np.sum(y_right + y_left)
    return T
```

##Método de Simpson

```python
import numpy as np

def simps(f,a,b,N):

    if N % 2 == 1:
        raise ValueError("N must be an even integer.")
    dx = (b-a)/N
    x = np.linspace(a,b,N+1)
    y = f(x)
    S = dx/3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])
    return S
```
